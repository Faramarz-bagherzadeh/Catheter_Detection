```python
import glob
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('train.csv')
data.head(2)
```

```python
data2=pd.DataFrame(columns=['UID','CLASS_1_ETT','CLASS_2_NGT''CLASS_3_CVC'])
data2['UID']=data['StudyInstanceUID']
data2.head()
```

```python
for UID in data2['UID']:
    _row=data[data['StudyInstanceUID']==UID]
    print(_row.index)
    if _row['ETT - Abnormal'].values[0]==1 or _row['ETT - Normal'].values[0]==1 or _row['ETT - Borderline'].values[0]==1:
        data2.loc[data2['UID']==UID,'CLASS_1_ETT']=1
    else:
        data2.loc[data2['UID']==UID,'CLASS_1_ETT']=0
        
    if _row['NGT - Abnormal'].values[0]==1 or _row['NGT - Normal'].values[0]==1 or _row['NGT - Borderline'].values[0]==1 or _row['NGT - Incompletely Imaged'].values[0]==1:
        data2.loc[data2['UID']==UID,'CLASS_2_NGT']=1
    else:
        data2.loc[data2['UID']==UID,'CLASS_2_NGT']=0
        
    if _row['CVC - Abnormal'].values[0]==1 or _row['CVC - Normal'].values[0]==1 or _row['CVC - Borderline'].values[0]==1:
        data2.loc[data2['UID']==UID,'CLASS_3_CVC']=1 
    else:
        data2.loc[data2['UID']==UID,'CLASS_3_CVC']=0
```
